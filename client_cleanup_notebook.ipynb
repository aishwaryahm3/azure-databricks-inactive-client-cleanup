## Applying filters

```python
#counting the total number of alive and dead people
df.groupBy("is_live").count().display()
```

```python
#counting people based on active status
df.groupBy('status').count().display()
```

```python
#checking people who are not active for more than 180 days
df.withColumn('active_in_180_days',when(col("last_updated") <= date_sub(current_date(), 180), "inactive").otherwise("active"))\
.withColumn('total_days',datediff(current_date(), col("last_updated")))\
.select('last_updated', 'total_days', 'active_in_180_days').display()
```

```python
#Let's query this all together
#Clean clients where:
#is_live != 'N'
#status NOT IN ('Inactive', 'Offboarded')
#last_updated within last 180 days

df_filtered = df.filter(
    (col("is_live") != "N") &
    (~col("status").isin("Inactive", "Offboarded")) &
    (datediff(current_date(), col("last_updated")) <= 180)
)
```

```python
#printing new df
df_filtered.display()
```

## Dropping Unused Cols

```python
cols_to_drop = ["email", "phone"]
df_cleaned = df_filtered.drop(*cols_to_drop)
```
